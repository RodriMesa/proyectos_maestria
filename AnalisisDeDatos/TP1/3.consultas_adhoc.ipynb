```python
import duckdb
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
with duckdb.connect("modelo_estrella.duckdb", read_only=True) as con:
   
    histograma_horas = con.sql(
        """
        SELECT HOUR(fecha_origen_recorrido) as  hora, WEEKDAY(fecha_origen_recorrido) AS dia_semana, id_estacion_origen
        FROM  fact_recorridos

        """
    ).df()
```

```python
# Set figure size before creating the plot


# Create your seaborn plot

plt.show()
```

```python
plt.figure(figsize=(13, 8))  # width=10 inches, height=6 inches
sns.countplot(
    data=histograma_horas,
    x="hora",
    hue="dia_semana",
    dodge=True,  # This creates the side-by-side effect
    palette="viridis"  # You can choose any palette
)

plt.title("Count of Events by Hour and Day of Week")
plt.xlabel("Hour of Day")
plt.ylabel("Count")
plt.legend(title="Day of Week")
plt.tight_layout()
plt.show()
#sns.barplot(x='category', y='values', data=df)
```
